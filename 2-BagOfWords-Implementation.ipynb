```python
import string
```

```python
def rem_punc(text): 
    return text.translate(str.maketrans('' , '' , string.punctuation)) 

df['text'] = df['text'].apply(rem_punc)
```

```python
#now removing numbers
```

```python
def rem_num(text): 
    new = ''
    for i in text: 
        if not i.isdigit():
            new = new+ i 
    return new 

df['text'] = df['text'].apply(rem_num)
```

```python
# now dealing with emojis or special characters we can use ascii values to identify them and then remove them 
```

```python
#as very character has its ascii value in python which is understandable to computer
```

```python
#ascii values are for characters that already exists but the unicode are for characters like emojis
```

```python
def rem_emo(text): 
    new = ''
    for i in text: 
        if i.isascii():
            new = new+ i 
    return new 

df['text'] = df['text'].apply(rem_emo)
```

```python
#for tokenization and removing stop__words we will use function
```

```python
def remove(text): 
    words = word_tokenize(text) #works onlt if punkt is downloaded and if you cant find punkt use "text.split()"
    cleaned = []    
    for i in words: 
        if i not in stop__words: 
            cleaned.append(i)
    return ' '.join(cleaned)

df['text'] = df['text'].apply(remove)
```

```python
#errored so do, 
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

# download required nltk data
nltk.download('punkt')
nltk.download('stopwords')

# define stop words
stop_words = set(stopwords.words('english'))

# define function
def remove(text):
    words = word_tokenize(text)
    cleaned = [word for word in words if word.lower() not in stop_words]
    return ' '.join(cleaned)

# example dataframe
df2 = pd.DataFrame({'text': ['This is an example sentence for cleaning text data.']})

# apply the function
df['text'] = df['text'].apply(remove)
```

```python
from sklearn.feature_extraction.text import CountVectorizer


#example data

document = ['i love pizza' , 'pizza is the best' ,'i love pasta' ,'pasta is great']

vector = CountVectorizer()  #All the preprocessing also is inbuilt in this as parameters so you can done using parameters here but doing it explicitily is a good practice

x= vector.fit_transform(document) 
print(vector.get_feature_names_out())  #Order doesn.t matter bydefault it give Vocabulary list in Alphabetical order
print(x.toarray())
print(x)
```

```python
#With n-gram range illustration
```

```python
#example data
document = ['i love pizza' , 'pizza is the best' ,'i love pasta' ,'pasta is great']

vector = CountVectorizer(ngram_range=(1,2)) 

x= vector.fit_transform(document) 
print(vector.get_feature_names_out())  #Order doesn.t matter bydefault it give Vocabulary list in Alphabetical order
print(x.toarray())
print(x)
```
